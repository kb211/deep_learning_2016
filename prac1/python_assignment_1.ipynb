Neural networks and deep learning methods prefer the input variables to contain as raw data as possible. 
But in the vast majority of cases data need to be preprocessed. What type of preprocessing you would prefer to use for each activation function and why? For example, in the previous cell we used per-example mean substraction.

**Your Answer**: While the Sigmoid function has seen frequent use historically, it is rarely used anymore due to its drawbacks: (1) at either of the tails of the Sigmoid, gradients are almost zero thus killing the gradient during back-propagation. This can happen during training when weights increase or during initialization. (2) During back-propagation, weights will either consist of all positive or all negative numbers when the data consists of only positive numbers. This is unpractical behaviour for back-propagation, as we would like to see each weight back-propagate in seperate directions. ReLU has shown faster SGD convergence, arguable due to its linear, non-saturating form, and can be very easily implemented by thresholding the activation matrix at 0.

Here standardization is a more suitable option than per-example mean substraction, as the image data contains multiple dimensions (color-channels), likely with a different distribution for each color channels per image. Therefore it is sensible to perform standardization on each dimension for each image seperately.

When using the Sigmoid activation function, per-example mean substraction makes sense as the Sigmoid is centered around zero. When using ReLU we should do standardization.

## Section 2: Multinomial Logistic Regression [5 points]

In this section you will get started by implementing a linear classification model called [Multinomial Logistic Regression](http://ufldl.stanford.edu/wiki/index.php/Softmax_Regression). Later on you will extend this model to a neural network. You will train it by using the [mini-batch Stochastic Gradient Descent algorithm](http://sebastianruder.com/optimizing-gradient-descent/index.html#minibatchgradientdescent). You should implement how to sample batches, how to compute the loss, how to compute the gradient of the loss with respect to the parameters of the model and how to update the parameters of the model. 

You should get around 0.35 accuracy on the validation and test sets with the provided parameters.


```python
# DONT CHANGE THE SEED AND THE DEFAULT PARAMETERS. OTHERWISE WE WILL NOT BE ABLE TO CORRECT YOUR ASSIGNMENT!
# Seed
np.random.seed(42)

# Default parameters. 
num_iterations = 1500
val_iteration = 100
batch_size = 200
learning_rate = 1e-7
weight_decay = 3e+4
weight_scale = 0.0001

########################################################################################
# TODO:                                                                                #
# Initialize the weights W using a normal distribution with mean = 0 and std =         #
# weight_scale. Initialize the biases b with 0.                                        #
########################################################################################   
# Weight matrix W = KxN (K = number of classes, M = length input vector)
W = np.random.normal(0, weight_scale, [num_classes, X_train.shape[1]])
# bias vector of length K
b = np.zeros(num_classes)

#--------------------------------
# My functions
# Function hyp computes matrix H, with elements P(y=k|X;W), where X is the train data and y is
# corresponding train label.
def hyp(W, b, X, Y):
    H = np.zeros((num_classes, X.shape[0])) # create hypotheses matrix
    H = np.exp(np.dot(W, np.transpose(X))+b[:,None]) # calculate exp of W*X+b
    H = np.divide(H, np.sum(H, axis=0)) # normalize matrix H columns
    return H

# Function calc_softmax returns the softmax loss, which is -sum(log(hypotheses)) of all
# hypotheses corresponding to the labels in Y.
def calc_softmax(W, b, X, Y):
    H = hyp(W, b, X, Y) # get hypotheses matrix H
    return -(1./X.shape[0])*np.sum([np.log(H[j,i]) for i,j in enumerate(Y)])

# Function calc_accuracy computes accuracy by counting the number of images where the
# highest probability is given to the hypothesis that correspond to the label y divided
# by the total number of images.
def calc_accuracy(W, b, X, Y):
    H = hyp(W, b, X, Y)
    return np.sum([int(np.argmax(H[:,i])==Y[i])
                   for i in range(Y.shape[0])])/float(Y.shape[0])

# Function calc_derivatives returns both the derivative towards the biases and the weights
def calc_derivatives(W, b, X, Y):
    # Create one-hot vector with 1 for the correct class-label.
    one_hot = np.zeros((num_classes, X.shape[0])) # create matrix with zeros
    # Fill one-hot vector and transpose it.
    one_hot = np.transpose(np.array([[int(i==y) for i in range(num_classes)]
                                     for y in Y]))
    # Calculate derivative towards bias by substracting hypotheses from the one-hot vector,
    # and computing the element-wise multiplication with H and (1-H), as it is equal to the
    # derivative towards weights and bias.
    # Derivative towards the weights is the derivative towards both the weights and the bias
    # times the derivative of the weights and the bias towards the weights, which is equal
    # to doing the dot product with the train data.
    H = hyp(W, b, X, Y)
    # First calculate elem-wise multilication
    temp_der = np.multiply((one_hot-H),np.multiply(H,1-H))
    db = -np.mean(temp_der, axis=1) # calculate db
    dW = -np.dot(temp_der, X)+weight_decay*W # calculate dW
    return db, dW
    
#--------------------------------
########################################################################################
#                              END OF YOUR CODE                                        #
########################################################################################

train_loss_history = []
train_acc_history = []

val_loss_history = []
val_acc_history = []

for iteration in range(num_iterations):
    ########################################################################################
    # TODO:                                                                                #
    # Sample a random mini-batch with the size of batch_size from the train set. Put the   #
    # images to X_train_batch and labels to Y_train_batch variables.                       #
    ########################################################################################
    # get random indexes to create mini-batch with
    idx = np.random.choice(X_train.shape[0], size=batch_size, replace=False)
    X_train_batch = X_train[idx,:] # using idx, create X_train_batch
    Y_train_batch = Y_train[idx] # using idx, create Y_train_batch
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # TODO:                                                                                #
    # Compute the loss and the accuracy of the multinomial logistic regression classifier  #
    # on X_train_batch, Y_train_batch.                                                     #
    ########################################################################################    
    # calculate softmax loss on the train batch
    train_loss = calc_softmax(W, b, X_train_batch, Y_train_batch)
    # calculate accuracy on the train batch
    train_acc = calc_accuracy(W, b, X_train_batch, Y_train_batch)
    # As there are 10 possible classes, the accuracy of random guesses should be around .1
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # TODO:                                                                                #
    # Compute the gradients of the loss with the respect to the weights and biases. Put    #
    # them in dW and db variables.                                                         #
    ########################################################################################
    # Compute derivates of the loss towards biases and weights.
    db, dW = calc_derivatives(W, b, X_train_batch, Y_train_batch)
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # TODO:                                                                                #
    # Update the weights W and biases b using the Stochastic Gradient Descent update rule. #
    ########################################################################################
    W = W - learning_rate*dW
    b = b - learning_rate*db
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    if iteration % val_iteration == 0 or iteration == num_iterations - 1:
        ########################################################################################
        # TODO:                                                                                #
        # Compute the loss and the accuracy on the validation set.                             #
        ########################################################################################
        val_loss = calc_softmax(W, b, X_val, Y_val)
        val_acc = calc_accuracy(W, b, X_val, Y_val)
        ########################################################################################
        #                              END OF YOUR CODE                                        #
        ########################################################################################
        train_loss_history.append(train_loss)
        train_acc_history.append(train_acc)
        val_loss_history.append(val_loss)
        val_acc_history.append(val_acc)
        
        # Output loss and accuracy during training
        print("Iteration {0:d}/{1:d}. Train Loss = {2:.3f}, Train Accuracy = {3:.3f}".
              format(iteration, num_iterations, train_loss, train_acc))
        print("Iteration {0:d}/{1:d}. Validation Loss = {2:.3f}, Validation Accuracy = {3:.3f}".
              format(iteration, num_iterations, val_loss, val_acc))

########################################################################################
# TODO:                                                                                #
# Compute the accuracy on the test set.                                                #
########################################################################################
test_acc = calc_accuracy(W, b, X_test, Y_test)
########################################################################################
#                              END OF YOUR CODE                                        #
########################################################################################
print("Test Accuracy = {0:.3f}".format(test_acc))
```

**Your Answer**: One factor that helps determine a good batch size is the amount of noise in the data. When data is very noisy, one should have a larger batch-size, so you even out the variance when updating your gradients. Too big batch-sizes on the other hand decrease the time efficiency of learning, as during one big batch, you could calculate several steps using a smaller batch, increasing performance faster. When using GPU's, one should adapt the batch-size for the available GPU memory. As time efficiency is now less influenced by bigger batches, it is beneficial to increase the batch size so that one batch fills up the GPU-memory.
